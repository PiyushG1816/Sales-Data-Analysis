```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Superstore Sales.xls")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Product Base Margin'].fillna(data['Product Base Margin'].mean(), inplace = True)
data.isnull().sum()
```

# ORDER PRIORITY


```python
data['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Order Priority' , data=data)
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
#THE ORDERS HAVE DIFFERENT PRIORITIES.
#WE SHOULD WORK ON INCREASING THE PRIORITY OF PRODUCTS HAVING LESS PRIORITY.
```

# SHIP MODE

```python
data['Ship Mode'].value_counts()
```

```python
a =data['Ship Mode'].value_counts()
l =['Regular Air', 'Delivery Truck' ,'Express Air']
plt.pie(x=a,labels = l, autopct="%1.2f%%",startangle=50)
plt.legend()
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# HOW THE PRODUCTS ARE SHIPPED TO THE DESTINATION.
# REGULAR AIR PLAYS A MAJOR IN SHIPPING.
```

# Relation Between Ship Mode and Product Category

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Ship Mode',data =data,hue="Product Category")
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# HOW MUCH OF THE PRODUCTS ARE SHIPPED ON THE BASIS OF SHIPPING TYPES.
# OFFICE SUPPLIES ARE MAJORLY SHIPPED BY REGULAR AIR.
```

# PRODUCT CATEGORY

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Product Category',data =data)
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# OFFICE SUPPLIES ARE THE MOST DEMANDED PRODUCTS.
```

# Relation Between Office products and thieir Sub-Category

```python
plt.figure(figsize=(9,7))
sns.countplot(x='Product Category',data =data[data['Product Category'] == 'Office Supplies'],hue ="Product Sub-Category")
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# SCISSORS, RULERS AND TRIMMERS ARE THE LEAST SUPPLIED PRODUCTS IN OFFICE SUPPLIES.
# PAPER IS THE MOST SUPPLIED PRODUCT IN OFFICE SUPPLIES.
```

# Relation Between Technology products and thieir Sub-Category

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Product Category',data =data[data['Product Category'] == 'Technology'],hue ="Product Sub-Category")
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# COPIERS AND FAX ARE THE LEAST SUPPLIED PRODUCTS IN TECHNOLOGY.
# TELEPHONES AND COMMUNICATION ARE THE MOST SUPPLIED PRODUCTS IN TECHNOLOGY.
```

# Relation Between Furniture products and thieir Sub-Category

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Product Category',data =data[data['Product Category'] == 'Furniture'],hue ="Product Sub-Category")
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# OFFICE FURNISINGS ARE THE MOST SUPLLIED PRODUCTS IN FURNITURE.
# BOOKCASES ARE THE MOST SUPPLIED PRODUCT IN FURNITURE.
```

# Relation Between Product Category and Product Base Margin

```python
plt.figure(figsize=(5,5))
sns.barplot(x= "Product Category", y="Product Base Margin",data =data)
plt.show()
```

```python
# FROM THE ABOVE DATA WE ANLYSEZED THAT:
# FURNITURE HAS THE MOST MARGIN IN PRODUCTS.
```
